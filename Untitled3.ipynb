```python
"""
========================================
Generalized Learning Vector Quantization
========================================
This example shows how GLVQ classifies.
The plot shows the target class of each data point
(big circle) and which class was predicted (smaller circle). It also
shows the prototypes (black diamond) and their labels (small point inside the diamond).

"""
import numpy as np
import matplotlib.pyplot as plt

from sklearn_lvq import GlvqModel
from sklearn_lvq.utils import plot2d

from scipy.optimize import minimize


class GlvqModelFixed(GlvqModel):
    def _optimize(self, x, y, random_state):
        label_equals_prototype = y[np.newaxis].T == self.c_w_
        res = minimize(
            fun=lambda vs: self._optfun(variables=vs, training_data=x, label_equals_prototype=label_equals_prototype),
            jac=lambda vs: self._optgrad(variables=vs, training_data=x, label_equals_prototype=label_equals_prototype, random_state=random_state),
            method='l-bfgs-b',
            x0=self.w_.ravel(),  # Fix here
            options={'disp': self.display, 'gtol': self.gtol, 'maxiter': self.max_iter})
        self.w_ = res.x.reshape(self.w_.shape)
        self.n_iter_ = res.nit



nb_ppc = 100
print('GLVQ:')
toy_data = np.append(
    np.random.multivariate_normal([0, 0], np.eye(2) / 2, size=nb_ppc),
    np.random.multivariate_normal([5, 0], np.eye(2) / 2, size=nb_ppc), axis=0)
toy_label = np.append(np.zeros(nb_ppc), np.ones(nb_ppc), axis=0)

glvq = GlvqModelFixed(prototypes_per_class=5)
glvq.fit(toy_data, toy_label)
plot2d(glvq, toy_data, toy_label, 1, 'glvq')

print('classification accuracy:', glvq.score(toy_data, toy_label))
plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_ikmeans_2d(model: iKMeans, X: np.ndarray, y: np.ndarray):
    # Convert to numpy just in case
    X = np.asarray(X)
    y = np.asarray(y)

    assert X.shape[1] == 2, "Only works with 2D data"

    # Get cluster assignments
    cluster_ids, true_labels = model.get_sample_assignments()
    centers = np.vstack(model.cluster_centers_)
    cluster_labels = model.cluster_labels_

    fig, ax = plt.subplots(figsize=(8, 6))

    # Plot data points with class colors
    for cls in np.unique(y):
        mask = y == cls
        ax.scatter(X[mask, 0], X[mask, 1], label=f'Class {int(cls)}', alpha=0.6, s=40)

    # Plot cluster centers
    for i, center in enumerate(centers):
        ax.scatter(center[0], center[1], marker='D', s=200,
                   edgecolor='k', facecolor='none', linewidth=1.5)
        ax.text(center[0], center[1], f'C{i}', fontsize=9, ha='center', va='center')

    ax.set_title("iKMeans Cluster Centers and Data")
    ax.legend()
    plt.axis('equal')
    plt.grid(True)
    plt.tight_layout()
    return ax
    # plt.show()
```

```python
class CONNiKMeans(iKMeans):
    def step_pred_first_and_second(self, x) -> int:
        """Predict the label for a single sample.

        Parameters
        ----------
        x : np.ndarray
            Data sample.

        Returns
        -------
        int
            Cluster label of the input sample.

        """
        assert len(self.cluster_centers_) >= 0, "iKmeans module is not fit."

        T = self.get_activation(x.reshape((1,-1))).reshape((-1))
        c1_ = int(np.argmax(T))
        if len(T) > 1:
            T = list(T)
            T[c1_] = -np.inf
            c2_ = int(np.argmax(T))
        else:
            c2_ = 1
        return c1_, c2_
```

```python
np.random.seed(50)
nb_ppc = 100
toy_data = np.append(
    np.random.multivariate_normal([0, 0], np.eye(2) / 2, size=nb_ppc),
    np.random.multivariate_normal([3, 0], np.eye(2) / 2, size=nb_ppc), axis=0)
toy_label = np.append(np.zeros(nb_ppc), np.ones(nb_ppc), axis=0)

model = CONNiKMeans(K=10, epsilon=0)
model.partial_fit(toy_data, toy_label)

ax = plot_ikmeans_2d(model, toy_data, toy_label)
i = 38
ax.scatter([toy_data[i,0]], [toy_data[i,1]], marker="X", color="red")
plt.show()
```
